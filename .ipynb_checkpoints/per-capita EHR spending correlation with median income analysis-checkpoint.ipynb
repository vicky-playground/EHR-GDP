```python
df_income = df_income.drop(df_income.columns[[3,4,6,7,8,9,10,11,12,13]], axis=1)
df_income.head()
```

```python
df_2 = df_income.merge(df_EHR, how='inner', left_on=['County', 'Taxable Year'], right_on=['Business_1', 'Payment_Ye'])
df_2.head()
```
